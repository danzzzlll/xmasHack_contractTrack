```python
import pandas as pd
import glob
import json
import PyPDF2

def make_dataframe():
    files_list = glob.glob("*.pdf")
    with open("classes.json") as f:
        classes = json.loads(f.read())
    classes = {key.split('.')[0]: value.split('/')[1] for key, value in classes.items()}
    df = pd.DataFrame({'uid': [], 'label': [], 'text': []})
    
    for file in files_list:
        pdfFileObj = open(file, 'rb')
        pdfReader = PyPDF2.PdfFileReader(pdfFileObj)
        pages_count = pdfReader.numPages
        text = ''
        for i in range(pages_count):
            pageObj = pdfReader.getPage(i)
            text += '\n' + pageObj.extractText()
        pdfFileObj.close()
        uid = file.split('/')[-1].split('.')[0]
        label = classes[uid]
        df2 = pd.DataFrame({'uid': [uid], 'label': [label], 'text': [text]})
        df = df.append(df2, ignore_index = True)
    return df
```

```python
df = make_dataframe()
```

```python
def replace_text(text):
    text = text.replace('ДОГОВОР АРЕНДЫ', '', 1)
    text = text.replace('ДОГОВОР ПОДРЯДА', '', 1)
    text = text.replace('ДОГОВОР КУПЛИ -ПРОДАЖИ', '', 1)
    text = text.replace('ДОГОВОР ПОСТАВКИ', '', 1)
    
    text = text[:text.find('Прежде, чем подписать, узнайте...')]
    return text
```

```python
df['text'] = df['text'].apply(replace_text)
```
